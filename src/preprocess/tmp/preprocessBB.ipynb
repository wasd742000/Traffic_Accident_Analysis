## Tiền xử lý cột 'Accident Date', '1st Road Class', '1st Road Class & No'

- '1st Road Class' và '1st Road Class & No' không có gì để phải tiền xử lý cả
- 'Accident Date' được ép kiểu về Datetime và sắp xếp tăng dần từ 2018->2019

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data_2018_2019.csv')
df.head()
```

```python
df['Accident Date'] = pd.to_datetime(df['Accident Date'])
df.sort_values(by=['Accident Date'], ascending=True, inplace=True)
df.head()
```
